```python
# Gerekli kütüphaneleri içe aktarın
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Öğrencinin çalışma saatleri ve final sınavındaki notları temsil eden veri örnekleri
# Bu veriler sadece örnek amaçlıdır ve gerçek dünya verileri olmayabilir.
# Öğrencinin çalışma saatleri
X = np.array([2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)

# Final sınavındaki notlar
Y = np.array([60, 65, 70, 75, 80, 85, 90])

# Doğrusal regresyon modelini oluşturun
model = LinearRegression()

# Modeli eğitin
model.fit(X, Y)

# Eğitilmiş modeli kullanarak tahmin yapın
Y_pred = model.predict(X)

# Sonuçları görselleştirin
plt.scatter(X, Y, color='blue', label='Gerçek Veriler')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Doğrusal Regresyon Tahmini')
plt.xlabel('Çalışma Saatleri')
plt.ylabel('Final Sınav Notu')
plt.title('Doğrusal Regresyon Örneği')
plt.legend()
plt.show()

# Eğitilmiş modelin katsayılarına erişin
print('Eğim (Katsayı):', model.coef_[0])
print('Düzlem (Kesim noktası):', model.intercept_)
```

```python
# Gerekli kütüphaneleri içe aktarın
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Öğrencinin çalışma saatleri ve sınavı geçip geçmediği (1: geçti, 0: geçmedi)
X = np.array([2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
Y = np.array([0, 0, 0, 1, 1, 1, 1])

# Lojistik regresyon modelini oluşturun
model = LogisticRegression()

# Modeli eğitin
model.fit(X, Y)

# Eğitilmiş modeli kullanarak tahmin yapın
Y_pred = model.predict(X)

# Sonuçları görselleştirin
plt.scatter(X, Y, color='blue', label='Gerçek Veriler')
plt.plot(X, Y_pred, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=8, label='Lojistik Regresyon Tahmini')
plt.xlabel('Çalışma Saatleri')
plt.ylabel('Sınav Geçme Durumu')
plt.title('Lojistik Regresyon Örneği')
plt.legend()
plt.show()
```
